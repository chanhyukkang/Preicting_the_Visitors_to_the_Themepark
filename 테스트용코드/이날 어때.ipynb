```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

dataset = pd.read_csv('Mall_Customers.csv')
dataset2 = pd.read_csv('mushroomsshort.csv')
dataset3 = dataset2[:200]#row만큼 자르기

dataset4 = pd.concat([dataset, dataset3], axis=1)#column을 합하기.
dataset5 = pd.concat([dataset4, dataset4], axis=0)

# print(dataset3)
# print(dataset4)
# print(dataset5)
# for i in range(len(dataset.values[0])):
#     print(dataset.values[:, i])#list형으로 변환(numypy)
    
dataset5 = pd.get_dummies(dataset)#one hot encoding data로 변환
print(dataset5)

ndataset = dataset5.values
print(ndataset)

#minmaxscaler를 통한 normalization
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(dataset5)
dataset5 = min_max_scaler.transform(dataset5)
print(dataset5)

#시계열데이터...는 날짜와 입장객의 feature로만 나타내면 된다.

```
